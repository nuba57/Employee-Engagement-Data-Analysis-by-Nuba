**An Introduction**

Definition HR analytics is the process of collecting and analyzing Human Resource (HR) data in order to improve an organization’s workforce performance. The process can also be referred to as talent analytics, people analytics, or even workforce analytics. This method of data analysis takes data that is routinely collected by HR and correlates it to HR and organizational objectives. Doing so provides measured evidence of how HR initiatives are contributing to the organization’s goals and strategies.

**Need for Analysis**

Analyzing this dataset helps organizations understand employee morale and well-being by identifying patterns and changes in engagement, satisfaction, and work-life balance over time — enabling data-driven strategies to improve retention, productivity, and workplace culture.

**Meaning of Classifications**:


Engagement Score: Reflects how emotionally committed employees are to their work and organization.

Satisfaction Score: Indicates how content employees are with their roles, compensation, and work environment.

Work-Life Balance Score: Measures how well employees can balance job responsibilities with personal life.

Survey Date: Represents the time when each employee’s feedback was collected, useful for tracking trends.

**Attribute Name Meanings**:

Employee ID – Unique identifier assigned to each employee who participated in the survey.

Survey Date – The date when the employee’s engagement survey response was recorded.

Engagement Score – A numerical value showing the employee’s emotional commitment and motivation toward their work.

Satisfaction Score – A score representing how satisfied the employee feels about their job, management, and environment.

Work-Life Balance Score – A measure indicating how well the employee can balance professional duties with personal life.
